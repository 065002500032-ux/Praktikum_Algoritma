```python

def dapatkan_akhiran_ordinal(n):
    if n == 0:
        return "th"

    dua_digit_terakhir = n % 100
    if 10 <= dua_digit_terakhir <= 20:
        return "th"

    digit_terakhir = n % 10

    if digit_terakhir == 1:
        return "st"
    elif digit_terakhir == 2:
        return "nd"
    elif digit_terakhir == 3:
        return "rd"
    else:
        return "th"



print("Ordinal Number")
print("ketik 0 untuk menghentikan program")

while True:
    angka = int(input("masukkan angka: "))

    if angka == 0:
        print("terima kasih")
        break

    suffix = dapatkan_akhiran_ordinal(angka)

    print(f"{angka}{suffix}")
```

```python
print("Ordinal Number")
print("ketik 0 untuk menghentikan program")

def ordinal_suffix(n):
    if 10 <= n % 100 <= 20:  # khusus 11,12,13
        return "th"
    last = n % 10
    if last == 1:
        return "st"
    elif last == 2:
        return "nd"
    elif last == 3:
        return "rd"
    else:
        return "th"

while True:
    angka = int(input("masukkan angka: "))
    suffix = ordinal_suffix(angka)
    print((angka, suffix))

    if angka == 0:
        print("terima kasih telah menggunakan program saya")
        break
```
