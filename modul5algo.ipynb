```python
konversi_nilai = {
    "A": 4.0,
    "B": 3.0,
    "C": 2.0,
    "D": 1.5,
    "E": 1.25,
    "B-": 2.75
}

nilai_angka = []

while True:
    huruf = input("masukkan nilai: ").upper()
    if huruf == "":
        break
    if huruf in konversi_nilai:
        angka = konversi_nilai[huruf]
        print("nilai =", angka)
        nilai_angka.append(angka)
    else:
        print("nilai tidak dikenali, coba lagi!")

# menghitung rata-rata
if nilai_angka:
    rata_rata = sum(nilai_angka) / len(nilai_angka)
    print("rata - rata nya adalah:", rata_rata)
else:
    print("tidak ada nilai yang dimasukkan.")
```

```python
total = 0.0

while True:
    umur_input = input("masukkan umur: ")

    if umur_input == "":
        break

    umur = int(umur_input)

    if umur <= 2:
        print("Gratis")
        harga = 0.0
    elif umur <= 12:
        harga = 14.00
        print(f"Harga ${harga:.2f}")
    elif umur <= 64:
        harga = 23.00
        print(f"Harga ${harga:.2f}")
    else:
        harga = 18.00
        print(f"Harga ${harga:.2f}")

    total += harga
    print(f"Running total: {total:.2f}")

uang = float(input("masukkan jumlah uang: "))
kembalian = uang - total
print(f"Running kembalian: {kembalian:.2f}")
```
