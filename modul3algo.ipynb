```python
"""
Nama : Haekal ramadhan
NIM : 065002500032
"""
print("=== Program Klasifikasi Segitiga ===")
a = int(input("Masukkan sisi a: "))
b = int(input("Masukkan sisi b: "))
c = int(input("Masukkan sisi c: "))
if a + b > c or a + c > b or b + c > a:
    if a == b == c:
        print("Segitiga Sama Sisi")
    elif a == b or b == c or a == c:
        print("Segitiga Sama Kaki")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("Segitiga Siku-Siku")
    else:
        print("Segitiga Sembarang")
else:
    print("Bukan segitiga (tidak memenuhi syarat segitiga)")
```

```python
print("Program Perhitungan Diskon Pembelanjaan")


nama = input("Masukkan Nama : ")
nim = input("Masukkan NIM  : ")
total_belanja = int(input("Masukkan Total Belanja Anda : Rp. "))


pakai_kupon = input("Bayar pakai kupon? (y/n) : ")


if pakai_kupon.lower() == "y":
    kode = input("Masukkan kode kupon : ")
    if kode == "haekal":
        print("\nSELAMAT ANDA MENDAPATKAN DISKON 100%")
        potongan = total_belanja
        total_bayar = 0
    else:
        print("\n KODE SALAH / EXPIRED")
        potongan = 0
        total_bayar = total_belanja
        print("BONUS: Anda tetap bisa lanjut belanja walau kode salah :)")


else:

    if 40000 <= total_belanja <= 89999:
        diskon = 0.20
    elif 90000 <= total_belanja <= 189999:
        diskon = 0.40
    elif 190000 <= total_belanja <= 389999:
        diskon = 0.60
    elif total_belanja >= 390000:
        diskon = 0.80
    else:
        diskon = 0.0

    potongan = total_belanja * diskon
    total_bayar = total_belanja - potongan

    if diskon == 0:
        print("\n ANDA TIDAK MENDAPATKAN DISKON")
    else:
        print(f"\n---SELAMAT ANDA MENDAPATKAN DISKON {int(diskon * 100)}%---")

print("\n---Rincian Belanja---")
print(f"{nama} - {nim}")
print(f"Total belanja anda : Rp.{total_belanja:,.2f}")
print(f"Potongan harga     : Rp.{potongan:,.2f}")
print(f"Total Pembayaran   : Rp.{total_bayar:,.2f}")
print("\nTerima kasih sudah belanja :)")
```
