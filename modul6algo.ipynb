```python
def konversi(huruf='A'):
    if huruf == 'A':
        return 4
    elif huruf == 'B':
        return 3
    elif huruf == 'C':
        return 2
    elif huruf == 'D':
        return 1
    else:
        return 0


def rata_rata(n1, n2):
    return (n1 + n2) / 2


d1 = input("Masukkan nilai: ").upper()
n1 = konversi(d1)
print("Nilai =", n1)

d2 = input("Masukkan nilai: ").upper()
n2 = konversi(d2)
print("Nilai =", n2)

print("Rata-ratanya adalah:", rata_rata(n1, n2))
```

```python
print("Program ini akan menentukan jumlah hari dalam bulan tertentu")
print("Masukkan 0 untuk menghentikan program")

while True:
    bulan = int(input("Masukkan bulan (1-12): "))

    if bulan == 0:
        print("Terima kasih sudah menggunakan program saya, sampai berjumpa lagi.")
        break

    # Jumlah hari tiap bulan (kecuali Februari yang khusus)
    if bulan in [1, 3, 5, 7, 8, 10, 12]:
        print("31 hari dalam sebulan")
    elif bulan in [4, 6, 9, 11]:
        print("30 hari dalam sebulan")
    elif bulan == 2:
        tahun = int(input("Masukkan tahun (e.g., 2021): "))
        # Mengecek tahun kabisat
        if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
            print("29 hari dalam sebulan")
        else:
            print("28 hari dalam sebulan")
    else:
        print("Input bulan tidak valid, silakan coba lagi.")
```
