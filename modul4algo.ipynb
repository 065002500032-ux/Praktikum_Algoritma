```python
"""
menghitung jumlah hari dalam bulan dan tahun tertentu menggunakan while
Nama = Haekal ramadhan
NIM = 065002500032
"""
ulang = "y"

while ulang.lower() == "y":
    bulan = int(input("Masukkan nomor bulan (1-12): "))
    tahun = int(input("Masukkan tahun: "))

    if bulan == 1:
        nama_bulan = "Januari"
        hari = 31
    elif bulan == 2:
        nama_bulan = "Februari"
        if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
            hari = 29
        else:
            hari = 28
    elif bulan == 3:
        nama_bulan = "Maret"
        hari = 31
    elif bulan == 4:
        nama_bulan = "April"
        hari = 30
    elif bulan == 5:
        nama_bulan = "Mei"
        hari = 31
    elif bulan == 6:
        nama_bulan = "Juni"
        hari = 30
    elif bulan == 7:
        nama_bulan = "Juli"
        hari = 31
    elif bulan == 8:
        nama_bulan = "Agustus"
        hari = 31
    elif bulan == 9:
        nama_bulan = "September"
        hari = 30
    elif bulan == 10:
        nama_bulan = "Oktober"
        hari = 31
    elif bulan == 11:
        nama_bulan = "November"
        hari = 30
    elif bulan == 12:
        nama_bulan = "Desember"
        hari = 31
    else:
        print("Nomor bulan tidak valid!")
        nama_bulan = "-"
        hari = 0

    if hari != 0:
        print(f"\nBulan {nama_bulan} tahun {tahun} memiliki {hari} hari.\n")

    ulang = input("Apakah ingin menghitung lagi? (y/n): ")

print("Program selesai.")
```
